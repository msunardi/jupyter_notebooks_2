```python
# Find cycles in an array
def findCycles(array):
    """Where each value in the array is a pointer to an index of the array, find cycles in array
    
    >>> array = [1,2,3,1,5,7,4]
    >>> findCycles(array)
    True
    >>> array = [1,3,5,7,9]
    >>> findCycles(array)
    False
    """
    n = 0
    adjacency_list = {}
    for i in range(len(array)):
        if i not in adjacency_list.keys():
            adjacency_list[i] = []
        adjacency_list[i].append(array[i])
        n += 1
    
    print(adjacency_list)
    
    for index, adjacents in adjacency_list.items():
        visited = {}
        n += 1
        
        try:
            print("Start: {}".format(index))
            goto = array[index]
            while goto in array:
                print("Goto: {}".format(goto))
                if goto >= len(array):
                    break
                if goto not in visited.keys():
                    visited[goto] = 0
                visited[goto] += 1                
                if visited[goto] > 1:
                    print(visited.keys())
                    print("O(): {}".format(n))
                    return True                
                goto = array[goto]
                n += 1
        except TypeError as e:
            print(index,adjacents)
    print("O(): {}".format(n))
    return False

array = [3,2,3,7,9,4,2,6,5]
array = [1,2,3,1,5,7,4]
array = [1,3,5,7,9]
array = [1,2,3,4,4]
array = [1,4,0,8,2]
findCycles(array)
    
```

```python
# Print all decimal numbers with length n
def printN(n, numrange, num=''):
    if n == 1:
        for i in numrange:
            print(num+str(i))
    else:
        for i in numrange:
            printN(n-1, numrange, ''.join([str(num), str(i)]))

def printDecimal(n):
    printN(n,range(10))
    
def diceRoll(n):
    printN(n, range(1,7,1))
printDecimal(2)
# diceRoll(3)
```

```python
# Write all permutations of word
def permute(word, n=0, chosen=''):
    word = list(set(word))
    n = len(word)
    if n == 1:
        for w in word:
            print(chosen+w)
    else:
        for w in range(n):
            permute(word[:w]+word[w+1:], n-1, chosen+word[w])

permute('card')
```
